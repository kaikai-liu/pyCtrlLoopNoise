## MZI resonator with balaced detection
 - Coupling power
 $$\kappa = \sin{\frac{\pi \delta n L_c}{\lambda}}$$
 - Coupling rate for resonator
 $$\gamma_{ex} = \frac{c\log{1/(1-\kappa^2)}}{n_g L}$$


```python
phi_list = np.array([0, 1/4, 2/4, 3/4, 4/4])*np.pi
r_in = 1
r_ex = 1
ka2 = 0.5

dw = np.linspace(-10, 10, 100)

y = np.array([PDB_MZI_resonator(dw, r_in, r_ex, ka2, ka2, phi) for phi in phi_list])
legs = [f'{x:.0f}/4'+r'$\pi$' for x in phi_list*4/np.pi]

plt.figure()
plt.plot(dw, y.T)
plt.xlabel(r'$\Delta\nu$ (MHz)')
plt.ylabel(r'$y_1 - y_2$')
plt.ylim((-1.1, 1.1))
plt.legend(legs)
```
